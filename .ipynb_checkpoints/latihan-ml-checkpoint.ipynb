```python
## Menguji akurasi model menggunakan data test
model_logreg.score(x_test, y_test)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

## Import data
df = pd.read_csv('Mall_Customers.csv')
##  Tampilkan 5 data pertama
df.head()
```

```python
## ubah nama kolom
df = df.rename(columns={'Gender':'gender', 'Age':'age', 'Annual Income (k$)':'income', 'Spending Score (1-100)':'spending_score'})

## ubah atribut gender dari kategorik menjadi numerik
df['gender'].replace(['Female', 'Male'], [0,1], inplace=True)

## hapus atribut yang tidak digunakan (CustomerID)
df.drop(['CustomerID', 'gender'], axis=1, inplace=True)

## tampilkan 5 data pertama
df.head()
```

```python
## Menentukan jumlah K
from sklearn.cluster import KMeans

data = df

# buat list inertia
clusters = []
for i in range (1,11):
    km = KMeans(n_clusters=i).fit(data)
    clusters.append(km.inertia_)

# tampilkan plot
fig, ax = plt.subplots(figsize=(8,4))
sns.lineplot(x=list(range(1,11)), y=clusters, ax=ax)
ax.set_title('Elbow')
ax.set_xlabel('Clusters')
ax.set_ylabel('Inertia')
```

```python
## Melatih model
# Create Model
model_kmeans5 = KMeans(n_clusters=5)
model_kmeans6 = KMeans(n_clusters=6)

# Train Model
model_kmeans5.fit(data)
model_kmeans6.fit(data)

# Menambah Label pada data
data['labels_5'] = model_kmeans5.labels_
data['labels_6'] = model_kmeans6.labels_
```

```python
data
```

```python
## Tampilkan Plot
plt.figure(figsize=(8,4))
sns.scatterplot(data['income'], data['spending_score'], hue=data['labels_5'], palette=sns.color_palette('hls',5))
plt.title(' KMeans 5 cluster')
plt.show()
```

```python
import pandas as pd
## Impoort data csv
df = pd.read_csv('Salary_Data.csv')
## Tampilkan 5 data awal
df.head()
```

```python
## Memisahkan atribut dan label
x = df['YearsExperience']
y = df['Salary']

## Transform to other axis
# Hal ini dilakukan karen hanya terdapat 1 atribut
x = x[:,np.newaxis]
```

```python
## Melakukan Grid Search untuk mencari parameter terbaik
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
model_svr_2 = SVR()
# list parameter dan nilai yg akan dicoba
parameters = {
    'kernel': ['rbf', 'linear', 'sigmoid'],
    'C': [1000, 10000, 100000],
    'gamma':[0.5, 0.05, 0.005]
}
# grid search
grid_search = GridSearchCV(model_svr_2, parameters)
grid_search.fit(x,y)
# tampilkan parameter terbaik
grid_search.best_params_
```

```python
## Melatih model dengan parameter terbaik
model_svr_new = SVR(kernel='linear', C=10000, gamma=0.5)
model_svr_new.fit(x,y)
```

```python
## Tampilkan Grafik
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.plot(x, model_svr_new.predict(x))
plt.show()
```

```python
### CNN - Convolutional Neural Network ###
```

```python
import tensorflow as tf
from tensorflow.keras.optimizers import RMSprop
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
print(tf.__version__)
```

```python
import os
base_dir = 'messy-clean'
train_dir = os.path.join(base_dir,'train')
val_dir = os.path.join(base_dir,'val')
```

```python
## memastikan directory benar
os.listdir(train_dir)
```

```python
## directory untuk masing-masing label pada train dan validasi
train_clean_dir = os.path.join(train_dir, 'clean')
train_messy_dir = os.path.join(train_dir, 'messy')
val_clean_dir = os.path.join(val_dir, 'clean')
val_messy_dir = os.path.join(val_dir, 'messy')
```

```python
image_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    horizontal_flip=True,
    shear_range=0.2,
    fill_mode='nearest'
)
```

```python
train_generator = image_datagen.flow_from_directory(
    train_dir,
    target_size=(150, 150),
    batch_size=4,
    class_mode='binary'
)

val_generator = image_datagen.flow_from_directory(
    val_dir,
    target_size=(150, 150),
    batch_size=4,
    class_mode='binary'
)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

model = Sequential([
    Conv2D(32,(3,3), activation='relu', input_shape=(150,150,3)),
    MaxPooling2D(2,2),
    Conv2D(32,(3,3), activation='relu'),
    MaxPooling2D(2,2),
    Conv2D(32,(3,3), activation='relu'),
    MaxPooling2D(2,2),
    Conv2D(32,(3,3), activation='relu'),
    MaxPooling2D(2,2),
    Flatten(),
    Dense(512, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```python
model.compile(
    loss='binary_crossentropy', 
    optimizer=tf.optimizers.Adam(),
    metrics=['accuracy'])
```

```python
model.fit(
    train_generator,
    steps_per_epoch=25,
    epochs=20,
    validation_data=val_generator,
    validation_steps=5,
    verbose=2
)
```
